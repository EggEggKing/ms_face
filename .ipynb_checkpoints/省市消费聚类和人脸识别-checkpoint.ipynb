## 机器学习的基本概念
- 机器学习分为：监督学习、无监督学习、强化学习、深度学习  
- 监督学习（有指导学习）： 数据有标签的，有结果的。例如：分类和回归  
- 无监督学习（无指导学习）： 数据没有标签，没有结果的。例如：聚类  
- 强化学习  
- 深度学习： 多层神经网络
